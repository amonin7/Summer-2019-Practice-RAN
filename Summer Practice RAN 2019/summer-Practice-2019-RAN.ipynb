# Random distribution images

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rnd_dst_data = []
with open("rand_distribution.txt") as f:
    for line in f:
        rnd_dst_data.append([float(x) for x in line.split()])
```

```python
rnd_dst_data.sort()
rnd_dst_data = np.array(rnd_dst_data)
```

```python
incVec = rnd_dst_data[:, 0]
incVec = incVec[::100]
decVec = rnd_dst_data[:, 1]
decVec.sort()
decVec = decVec[::100]
z = rnd_dst_data[:, 2]
```

```python
from mpl_toolkits.axes_grid1 import make_axes_locatable

VMIN = 0
VMAX = 750
STEP = 75

def plot_color_map(z, alpha_values, tau_values, alpha_freq= 15, tau_freq=10, dist_type=""):
    z = np.array(z)
    Z = z.reshape((len(tau_values), len(alpha_values)))

    plt.figure(figsize=(5, 5))
    ax = plt.gca()

    plt.title("ROZ func 2 vars. with distribution type: " + dist_type)
    plt.xlabel("DECREMENT", fontsize=12)
    plt.ylabel("INCREMENT", fontsize=12)
    plt.xticks([alpha_freq*i for i in range(len(alpha_values[::alpha_freq]))], np.round(alpha_values[::alpha_freq], decimals=3))
    plt.yticks([tau_freq*i for i in range(len(tau_values[::tau_freq]))], [np.round(el, decimals=2) for el in tau_values[::tau_freq]])

    im = ax.imshow(Z, interpolation='bilinear', cmap='winter', aspect='auto', vmin=VMIN, vmax=VMAX)

    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.5)

    cbar = plt.colorbar(im, cax=cax)
    cbar.set_ticks(np.arange(VMIN, VMAX, STEP))
    cbar.set_ticklabels(np.arange(VMIN, VMAX, STEP))
    plt.show()
```

```python
plot_color_map(z, decVec, incVec)
```

### Plotting the found minimum of the function (random distribution)

```python
data = []
with open("rand_distribution.txt") as f:
    for line in f:
        data.append([float(x) for x in line.split()])
data = np.array(data)
values = data[1:10000, 3]
x = [i for i in range(1, 10000, 1)]
plt.figure(figsize=(15,5))
plt.plot(x, values)
plt.show()
```
